```python
# import all relevant modules

from citipy import citipy
import requests
import random
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
# I can do this command:
# from config import api_key
# and saved a one liner config.py file in the same folder as this notebook
# to save time I am not doing that 

# weather map key for Tushar Makhija is in the notes section of bootcampspot
api_key = "see bootcampspot notes"

# number of random cities to generate
# I kept a small number here to start off, HW instructions want it to be 500
num_city = 500
```

```python
#generate random 500 unique cities
city_500 = []

while len(set(city_500)) <num_city:

    # lat and lng randomized
    lat = random.random()*90 *random.choice([-1,1])
    lng = random.random()*180 *random.choice([-1,1])

    # random city name
    city = citipy.nearest_city(lat, lng)
    city_500.append(city.city_name)

city_500 = list(set(city_500))
len(city_500)
```

```python
#placeholders for the data I will need from the JSON

lat = []
current_temp_degreesF = []
current_humidity = []
current_cloudiness = []
current_windspeed_mph = []
```

```python
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"

for cityname in city_500:
    print("processing {}".format(cityname))
    
    # Build query URL
    query_url = url + "appid=" + api_key + "&q=" + cityname

    # Get weather data
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()

    try:
        # append the data to the respective lists
        lat.append(weather_json["coord"]["lat"])

        # converted from Kelvin to farenheit and append
        current_temp_degreesF.append(1.8 * ( weather_json["main"]["temp_max"] - 273) + 32) 

        current_humidity.append(weather_json["main"]["humidity"])

        current_cloudiness.append(weather_json["clouds"]["all"])

        # converted from m/s to mph and append
        current_windspeed_mph.append(weather_json["wind"]["speed"] * 2.23694)
    
    except KeyError:
        
        print("could not access {}, onto the next one".format(cityname))
        


```

```python
# convert lists to df
# this df is what I will write to csv later

df = pd.DataFrame({"lat": lat, "temperature":current_temp_degreesF,
                  "humidity": current_humidity,"cloudiness": current_cloudiness,
                  "windspeed": current_windspeed_mph})

df.head()
```

```python
#graph 1

sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(x = 'temperature', y = "lat", data=df, palette="plasma" )
plt.xticks(rotation='vertical')
plt.title("Temp vs Lattitude")
# plt.ylim(2500,5000)
plt.savefig("tempVlat.png")
plt.show()
```

```python
#graph 2

sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(x = 'humidity', y = "lat", data=df, palette="plasma" )
plt.xticks(rotation='vertical')
plt.title("Humidity vs Lattitude")

plt.savefig("humidityVlat.png")
plt.show()
```

```python
#graph 3

sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(x = 'cloudiness', y = "lat", data=df)
plt.xticks(rotation='vertical')
plt.title("Cloudiness vs Lattitude")
plt.savefig("cloudinessVlat.png")
plt.show()
```

```python
#graph 4

sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(x = 'windspeed', y = "lat", data=df)
plt.xticks(rotation='vertical')
plt.title("Windspeed vs Lattitude")
plt.savefig("windspeedVlat.png")
plt.show()
```

```python
# export df as .csv
df.to_csv("weather.csv", index = False)
```
